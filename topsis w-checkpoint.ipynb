```python
import pandas as pd
import numpy as np
import math
import sys
df = pd.read_csv('topsis1.csv')
df.head(20)
```

```python
w = [0.059246, 0.4828, 0.31385, 0.1441]
```

```python
data=df.drop(columns=["Alternatives"])
data
```

```python
X_norm = data/np.sqrt(np.power(data,2).sum(axis=0))
X_norm
```

```python
X_norm_w = X_norm*w
X_norm_w
```

```python
X_norm_w.min()
```

```python
X_norm_w.max()
```

```python
aaa = [X_norm_w.max()[0],X_norm_w.max()[1],X_norm_w.max()[2],X_norm_w.min()[3] ]
aaa
```

```python
bb = [X_norm_w.min()[0],X_norm_w.min()[1],X_norm_w.min()[2],X_norm_w.max()[3] ]
bb
```

```python
positive_ideal = aaa
negative_ideal = bb
```

```python
#positive_ideal
SM_P = np.sqrt(np.power(X_norm_w-positive_ideal,2).sum(axis=1))
#negative_ideal
SM_N = np.sqrt(np.power(X_norm_w-negative_ideal,2).sum(axis=1))
```

```python
SM_N
```

```python
Per_score = SM_N/(SM_P+SM_N)
Per_score
```
